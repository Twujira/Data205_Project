```python
# Get summary statistics for the '2020' column with 25th and 75th percentiles
summary = County_pop['2020'].describe(percentiles=[.25, .75])

# Convert the summary statistics to integers
summary_int = summary.astype(int)

# Get the mode for the '2020' column and convert to int
mode = int(County_pop['2020'].mode()[0])

# Displaying the summary
print(summary_int)
print(f"Mode: {mode}")
```
